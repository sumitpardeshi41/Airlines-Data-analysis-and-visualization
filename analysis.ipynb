Checking for Airlines in the Dataset, accompanied by their Frequencies 


```python
## Checking for number of unique Airlines in the dataset
df['airline'].nunique()
```

```python
df['airline'].unique()
```

```python
### Showing Frequencies of Each airline
df['airline'].value_counts()
```

```python
### Showing the frequenices of each airline using Horizontal bar graph
freq=df['airline'].value_counts(ascending=True)
freq.plot.barh(color=['pink','purple'])
plt.title('Airline Frequency Graph')
plt.xlabel('Number of Frequency')
plt.ylabel('Airline Company')
plt.show()
```

Checking if Price varies with Airlines

```python
### Grouping the airlines and checking their mean prices
df.groupby('airline')['price'].mean()
```

```python
## Drawing a Categorical plot showing the Mean Ticket Price for each airline

sns.catplot(x='airline',y='price',kind='bar',hue='class',legend=True,palette='rocket',data=df)
```
